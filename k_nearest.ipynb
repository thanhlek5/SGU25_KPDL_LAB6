```python
import pandas as pd 
from K_nearest_model import K_Nearest_Neighbors
import random
```

```python
features_name = ['lettr','x-box','y-box','width','high ','onpix','x-bar','y-bar','x2bar','y2bar','xybar','x2ybr','xy2br','x-ege','xegvy','y-ege','yegvx']
file_path = r'C:\Users\LENOVO\OneDrive\Desktop\khai_pha\LAB6\letter+recognition\letter-recognition.data'
df = pd.read_csv(file_path,header=None,names=features_name)
```

```python
def format_data_for_knn(data_list):
    """
    Chuyển đổi một list có dạng [['A', 1, 2], ['B', 3, 4], ['A', 5, 6]]
    thành một dict có dạng {'A': [[1, 2], [5, 6]], 'B': [[3, 4]]}
    
    Đây là định dạng mà class KNN của bạn yêu cầu.
    """
    data_dict = {}
    
    # Duyệt qua từng hàng trong list dữ liệu của bạn
    for row in data_list:
        group = row[0]      # Lấy nhãn (ví dụ: 'A', 'B', 'C'...)
        features = row[1:]    # Lấy các đặc trưng (tất cả trừ cột đầu tiên)
        
        # Nếu nhãn này chưa có trong dictionary, hãy tạo một list rỗng cho nó
        if group not in data_dict:
            data_dict[group] = []
            
        # Thêm list các đặc trưng vào đúng nhãn của nó
        data_dict[group].append(features)
        
    return data_dict
```

```python
data_list = df.values.tolist()
random.shuffle(data_list)

test_size = 0.2
total_samples = len(data_list)
split_index = int(total_samples * (1 - test_size))
train_list = data_list[:split_index]
test_list = data_list[split_index:]

# Dùng hàm này để chuyển đổi cả hai list
data_train = format_data_for_knn(train_list)
data_test = format_data_for_knn(test_list)


```

```python
print(f"Tổng số mẫu: {total_samples}")
print(f"Số mẫu Train (80%): {len(data_train)}")
print(f"Số mẫu Test (20%): {len(data_test)}")
```

```python
knn = K_Nearest_Neighbors(data_train,k = 7)
knn.test(data_test)


```
